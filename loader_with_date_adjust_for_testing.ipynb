```python
import os
import pandas as pd
from functools import reduce
raw_files_path = '/content/gdrive/MyDrive/raw_data/'
flat_files_path = '/content/gdrive/MyDrive/test_flat_data/'
```

```python

raw_files = os.listdir(raw_files_path)

i = 0
for raw_file in raw_files:
  i = i + 1
  if i != 2 and i != 523 and i != 1109:
    continue
  if os.path.exists(flat_files_path + raw_file):
    print(raw_file + ' already processed!', ';' + str(len(raw_files) - i) + ' left')
    continue
  try:
    xls = pd.ExcelFile(raw_files_path + raw_file)
    df_Prov_CF_Con = pd.read_excel(xls, 'Prov_CF_Con')

    df_Prov_CF_Con['DISCL_DATE'] = pd.to_datetime(df_Prov_CF_Con['Period'], format='%Y%m')
    df_Prov_CF_Con['DISCL_DATE'] = df_Prov_CF_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    print(df_Prov_CF_Con[['Compid','Period','DISCL_DATE']].head)

    df_Prov_CF_SA = pd.read_excel(xls, 'Prov_CF_SA')

    df_Prov_CF_SA['DISCL_DATE'] = pd.to_datetime(df_Prov_CF_SA['Period'], format='%Y%m')
    df_Prov_CF_SA['DISCL_DATE'] = df_Prov_CF_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    df_Prov_BS_Con = pd.read_excel(xls, 'Prov_BS_Con')

    df_Prov_BS_Con['DISCL_DATE'] = pd.to_datetime(df_Prov_BS_Con['Period'], format='%Y%m')
    df_Prov_BS_Con['DISCL_DATE'] = df_Prov_BS_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    df_Prov_BS_SA = pd.read_excel(xls, 'Prov_BS_SA')

    df_Prov_BS_SA['DISCL_DATE'] = pd.to_datetime(df_Prov_BS_SA['Period'], format='%Y%m')
    df_Prov_BS_SA['DISCL_DATE'] = df_Prov_BS_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    df_Qtr_PnL_Con = pd.read_excel(xls, 'Qtr_PnL_Con')

    df_Qtr_PnL_Con['DISCL_DATE'] = pd.to_datetime(df_Qtr_PnL_Con['Period'], format='%Y%m')
    df_Qtr_PnL_Con['DISCL_DATE'] = df_Qtr_PnL_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    df_Qtr_PnL_SA = pd.read_excel(xls, 'Qtr_PnL_SA')

    df_Qtr_PnL_SA['DISCL_DATE'] = pd.to_datetime(df_Qtr_PnL_SA['Period'], format='%Y%m')
    df_Qtr_PnL_SA['DISCL_DATE'] = df_Qtr_PnL_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(40))

    df_PnL_Con = pd.read_excel(xls, 'PnL_Con')

    df_PnL_Con['DISCL_DATE'] = pd.to_datetime(df_PnL_Con['Period'], format='%Y%m')
    df_PnL_Con['DISCL_DATE'] = df_PnL_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    print(df_PnL_Con[['Compid','Period','DISCL_DATE']].head)

    df_PnL_SA = pd.read_excel(xls, 'PnL_SA')

    df_PnL_SA['DISCL_DATE'] = pd.to_datetime(df_PnL_SA['Period'], format='%Y%m')
    df_PnL_SA['DISCL_DATE'] = df_PnL_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    df_CFS_Con = pd.read_excel(xls, 'CFS_Con')

    df_CFS_Con['DISCL_DATE'] = pd.to_datetime(df_CFS_Con['Period'], format='%Y%m')
    df_CFS_Con['DISCL_DATE'] = df_CFS_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    df_CFS_SA = pd.read_excel(xls, 'CFS_SA')

    df_CFS_SA['DISCL_DATE'] = pd.to_datetime(df_CFS_SA['Period'], format='%Y%m')
    df_CFS_SA['DISCL_DATE'] = df_CFS_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    df_BSS_Con = pd.read_excel(xls, 'BSS_Con')

    df_BSS_Con['DISCL_DATE'] = pd.to_datetime(df_BSS_Con['Period'], format='%Y%m')
    df_BSS_Con['DISCL_DATE'] = df_BSS_Con['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    df_BSS_SA = pd.read_excel(xls, 'BSS_SA')

    df_BSS_SA['DISCL_DATE'] = pd.to_datetime(df_BSS_SA['Period'], format='%Y%m')
    df_BSS_SA['DISCL_DATE'] = df_BSS_SA['DISCL_DATE'].apply(lambda x: x + pd.offsets.Day(150))

    dfs = [df_Prov_CF_Con, df_Prov_CF_SA, df_Prov_BS_Con, df_Prov_BS_SA, df_Qtr_PnL_Con, df_Qtr_PnL_SA, df_PnL_Con, df_PnL_SA, df_CFS_Con, df_CFS_SA, df_BSS_Con, df_BSS_SA]
    df_all = reduce(lambda left,right: pd.merge(left, right, how='outer', left_on=['Compid','Period','DISCL_DATE'], right_on=['Compid','Period','DISCL_DATE']),dfs)

    #df_all = pd.merge([df_Prov_CF_Con, df_Prov_CF_SA, df_Prov_BS_Con, df_Prov_BS_SA, df_Qtr_PnL_Con, df_Qtr_PnL_SA, df_PnL_Con, df_PnL_SA, df_CFS_Con, df_CFS_SA, df_BSS_Con, df_BSS_SA], axis=1, join='outer', ignore_index=True, on=['Compid','Period','DISCL_DATE'])
    print('after merge')
    print(df_all[['Compid','Period','DISCL_DATE']].head)    
    df_all = df_all.loc[:,~df_all.columns.duplicated()]
    print('after dedupe')
    print(df_all[['Compid','Period','DISCL_DATE']].head)
    df_all.to_csv(flat_files_path + raw_file, encoding='utf-8', index=False)
    print('processed ' + raw_file, ';' + str(len(raw_files) - i) + ' left')
  except Exception as e:
    print('issue with ', raw_file, ' ', e)
```
